```python
#Import libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
import numbers as np
```

```python
# link variable with the search
url = 'https://www.imdb.com/search/name/?birth_place=Spain&adult=include&count=100&start=1&ref_=rlm'
```

```python
# Returns an HTML from the url and a fixed width of 60 pixels.
def path_to_image_html(url):
    return '<img src="'+ url + '" width="60" >'
```

```python
# Fetches the text content from the specified URL and stores it in the 'webtxt' variable
webtxt = requests.get(url).text
print(webtxt)
```

```python
# Parses the HTML content ('webtxt') and creates an object for further processing
soup = BeautifulSoup(webtxt,'lxml')
print(soup)
```

```python
# Print the content in a more readable structured format.
print(soup.prettify())
```

```python
# Checking the titles.
print(soup.title.string)
```

```python
# Find and count 'actors' using specified class.
actors = soup.find_all('div', class_ = 'lister-item mode-detail')
print(type(actors))
print(len(actors))
```

```python
# Convert the list of actors to a string.
html_string = (str(actors))

# Parse the html_string.
soup = BeautifulSoup(html_string, 'html.parser')

# Create empty lists to store data.
names = []
image = []
bios = []

# Find all actor elements in the HTML.
actors = soup.find_all('div', class_ = 'lister-item mode-detail')


for actor in actors:
    
# Extract and append the actor's name.
    name = actor.h3.a.text.strip()
    names.append(name)
    
# Extract and append the actor's image URL.    
    picture = actor.find('img')['src']
    image.append(picture)
    
# Extract and append a shortened bio.
    bio = actor.p.text.strip()
    bio = actor.text.split(bio)[1].strip()
    bio = list(bio.split("."))[0]
    bios.append(bio)

# Create a DataFrame with the extracted data.
df = pd.DataFrame({'Actor Name': names, 'Picture': image, 'Bios': bios})
df
```

```python
# Import necessary libraries to display HTML content.
from IPython.core.display import HTML

# Function to create HTML image tags for displaying images.
def display_image(url):
    return f'<img src="{url}" width="60" >'

# Apply the 'display_image' function to the 'Picture' column of the DataFrame.
df['Picture'] = df['Picture'].apply(display_image)

# Display the DataFrame as an HTML table with images.
display(HTML(df.to_html(escape=False, render_links=True)))
```
